```python

# Load the IMDB Word Index
word_index = imdb.get_word_index()

#  Print a Sample of the Word Index (First 10 Items)
print("Total words in dictionary:", len(word_index))
print("Sample words:", list(word_index.items())[:10])  # Display first 10 words

# Reverse the Word Index to Decode Reviews
reverse_word_index = {value: key for key, value in word_index.items()}

# Load the IMDB Dataset
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

# Decode a Sample Review (First Review in Training Set)
decoded_review = ' '.join([reverse_word_index.get(i - 3, '?') for i in train_data[0]])

# Display the First Decoded Review
print("\nSample Decoded Review:")
print(decoded_review)
```
